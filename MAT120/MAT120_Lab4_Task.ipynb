```python
import numpy as np
import matplotlib.pyplot as plt

# Define Lotka-Volterra equations
def dR_dt(R, W):
    return R * (0.08 - 0.001 * W)

def dW_dt(R, W):
    return W * (-0.02 + 0.00002 * R)

# Initial conditions
R_initial = 1000  # Initial population of rabbits
W_initial = 40    # Initial population of wolves
t_initial = 0     # Initial time
t_final = 100     # Final time
dt = 0.1          # Time step size

# Lists to store population values and time
R_values = [R_initial]
W_values = [W_initial]
time_values = [t_initial]

# Euler's method
while time_values[-1] < t_final:
    R = R_values[-1]
    W = W_values[-1]
    dRdt = dR_dt(R, W)
    dWdt = dW_dt(R, W)
    R_new = R + dRdt * dt
    W_new = W + dWdt * dt
    R_values.append(R_new)
    W_values.append(W_new)
    time_values.append(time_values[-1] + dt)

# Plotting the results
plt.figure(figsize=(15, 5))

# Plot W vs t
plt.subplot(1, 3, 1)
plt.plot(time_values, W_values, 'b-')
plt.xlabel('Time')
plt.ylabel('Wolves Population')
plt.title('Wolves Population vs Time')

# Plot R vs t
plt.subplot(1, 3, 2)
plt.plot(time_values, R_values, 'g-')
plt.xlabel('Time')
plt.ylabel('Rabbits Population')
plt.title('Rabbits Population vs Time')

# Plot W vs R
plt.subplot(1, 3, 3)
plt.plot(R_values, W_values, 'r-')
plt.xlabel('Rabbits Population')
plt.ylabel('Wolves Population')
plt.title('Wolves Population vs Rabbits Population')

plt.tight_layout()
plt.show()
```
