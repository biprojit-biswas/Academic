```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the function
def func(x):
    return -2*x**3 + 16*x**2 - 15*x

# Define the range
x = np.linspace(-10, 10, 400)
y = func(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = -2x^3 + 16x^2 - 15x', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the function')
plt.grid(True)
plt.legend()

# Find minima and maxima
result_min = minimize(func, -10)
result_max = minimize(lambda x: -func(x), -10)
if result_min.success:
    min_x = result_min.x[0]
    min_y = result_min.fun
    plt.plot(min_x, min_y, 'ro', label=f'Minima: ({min_x:.2f}, {min_y:.2f})')
    plt.text(min_x, min_y, f'Minima: ({min_x:.2f}, {min_y:.2f})', verticalalignment='bottom')
else:
    print("No minima found in the given range.")

if result_max.success:
    max_x = result_max.x[0]
    max_y = -result_max.fun
    plt.plot(max_x, max_y, 'go', label=f'Maxima: ({max_x:.2f}, {max_y:.2f})')
    plt.text(max_x, max_y, f'Maxima: ({max_x:.2f}, {max_y:.2f})', verticalalignment='bottom')
else:
    print("No maxima found in the given range.")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.log(x)

# Define the interval
x_values = np.linspace(np.pi, 5*np.pi, 500)
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values, label='y = ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = ln(x)')
plt.grid(True)
plt.legend()

# Find minima and maxima
minima = np.min(y_values)
maxima = np.max(y_values)

minima_location = x_values[np.argmin(y_values)]
maxima_location = x_values[np.argmax(y_values)]

print("Minimum value of y within the interval: ", minima)
print("Location of minimum: x =", minima_location)
print("Maximum value of y within the interval: ", maxima)
print("Location of maximum: x =", maxima_location)

# Highlight minima and maxima points on the graph
plt.scatter([minima_location, maxima_location], [minima, maxima], color='red')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.sin(2*x)

# Define the range
x = np.linspace(-np.pi, np.pi, 1000)
y = f(x)

# Plot the function
plt.plot(x, y, label='y = sin(2x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(2x) over [-π, π]')
plt.grid(True)
plt.legend()
plt.show()

# Finding minima and maxima
minima_x = []
maxima_x = []
for i in range(1, len(x)-1):
    if y[i] < y[i-1] and y[i] < y[i+1]:
        minima_x.append(x[i])
    elif y[i] > y[i-1] and y[i] > y[i+1]:
        maxima_x.append(x[i])

if minima_x:
    print("Minima at x =", minima_x)
else:
    print("No minima in the given interval.")

if maxima_x:
    print("Maxima at x =", maxima_x)
else:
    print("No maxima in the given interval.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * np.cos(x) + 3 * np.sin(x)


# Generate x values
x = np.linspace(0, 15, 1000)

# Generate y values
y = f(x)

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = 2 \cos(x) + 3 \sin(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = 2cos(x) + 3sin(x)')
plt.grid(True)
plt.legend()
plt.show()

# Find minima and maxima
extrema_x = []
extrema_y = []
for i in range(1, len(x) - 1):
    if y[i] > y[i - 1] and y[i] > y[i + 1]:
        extrema_x.append(x[i])
        extrema_y.append(y[i])
    elif y[i] < y[i - 1] and y[i] < y[i + 1]:
        extrema_x.append(x[i])
        extrema_y.append(y[i])

print("Minima and Maxima:")
if len(extrema_x) == 0:
    print("No minima or maxima in the given interval.")
else:
    for i in range(len(extrema_x)):
        print(f"At x = {extrema_x[i]}, y = {extrema_y[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.tan(x / 4)

# Define the range
x_values = np.linspace(-3 * np.pi, 3 * np.pi, 1000)
y_values = f(x_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$y = \tan\left(\frac{x}{4}\right)$')
plt.title('Graph of y = tan(x/4)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Find local minima and maxima within the given interval
local_minima = []
local_maxima = []

for i in range(1, len(y_values) - 1):
    if y_values[i] < y_values[i - 1] and y_values[i] < y_values[i + 1]:
        local_minima.append((x_values[i], y_values[i]))
    elif y_values[i] > y_values[i - 1] and y_values[i] > y_values[i + 1]:
        local_maxima.append((x_values[i], y_values[i]))

# Display minima and maxima
print("Local Minima:")
if len(local_minima) == 0:
    print("No local minima found in the given interval.")
else:
    for minimum in local_minima:
        print("x =", minimum[0], ", y =", minimum[1])

print("\nLocal Maxima:")
if len(local_maxima) == 0:
    print("No local maxima found in the given interval.")
else:
    for maximum in local_maxima:
        print("x =", maximum[0], ", y =", maximum[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the function
def func(x):
    return 10*x**3 + 9*x**2 - 7*x - 5

# Define the interval
x_values = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_values = func(x_values)

# Plot the function
plt.plot(x_values, y_values, label='y = 10x^3 + 9x^2 - 7x - 5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the function')
plt.grid(True)
plt.legend()
plt.show()

# Find minima and maxima using optimization
minima_result = minimize(func, 0)
maxima_result = minimize(lambda x: -func(x), 0)

if minima_result.success:
    minima_x = minima_result.x[0]
    minima_y = minima_result.fun
    print(f"Minimum at (x, y) = ({minima_x}, {minima_y})")
else:
    print("No minimum found in the given interval.")

if maxima_result.success:
    maxima_x = maxima_result.x[0]
    maxima_y = -maxima_result.fun
    print(f"Maximum at (x, y) = ({maxima_x}, {maxima_y})")
else:
    print("No maximum found in the given interval.")
```

```python
import sympy as smp
# Define symbols and constants
x = smp.symbols('x', real=True)
b = smp.Symbol('b', positive=True)
a = smp.Symbol('a', positive=True)

#f= smp.exp(-x)*(x**2)*smp.cos(x)

f= smp.exp(-b*x) * smp.sin(a*x)

# Integrate symbolically
integral = smp.integrate(f, x)

print("Symbolic Integral:", integral)
smp.integrate(f,x).simplify()
```
