```python
#Task3

import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        print(f'Vector <{self.x}, {self.y}, {self.z}> created.')

    def magnitude(self):
        mag = math.sqrt(self.x**2 + self.y**2 + self.z**2)
        return mag

    def dot_product(self, other):
        dot = self.x*other.x + self.y*other.y + self.z*other.z
        return dot

    def cross_product(self, other):
        x = self.y*other.z - self.z*other.y
        y = self.z*other.x - self.x*other.z
        z = self.x*other.y - self.y*other.x
        return Vector3D(x, y, z)



v1 = Vector3D(2, -3, 1)
v2 = Vector3D(-1, 4, 0)

print(f'Magnitude of the first vector = {v1.magnitude()}')
print(f'Magnitude of the second vector = {v2.magnitude()}')

dot_product = v1.dot_product(v2)
print(f'Dot product of the two vectors = {dot_product}')

cross_product = v1.cross_product(v2)
print(f'Cross product of the two vectors = <{cross_product.x}, {cross_product.y}, {cross_product.z}>')
```

```python
#Task4

class abcTech:
    def __init__(self, name, designation, department):
        self.name = name
        self.designation = designation
        self.department = department
        self.programming_skills = []
        self.frameworks = []
        print(f"Welcome to abcTech, {self.name}")

    def addProgrammingSkills(self, skills):
        self.programming_skills.extend(skills)

    def addFrameworks(self, frameworks):
        self.frameworks.extend(frameworks)

    def printInfo(self):
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Department: {self.department}")
        print(f"Programming Skills: {', '.join(self.programming_skills)}")
        print(f"Frameworks: {', '.join(self.frameworks)}")

    def calculateSalary(self, base_salary, working_hours):
        if working_hours > 144:
            extra_hours = working_hours - 144
            extra_pay = extra_hours * 800
            total_pay = base_salary + extra_pay
        else:
            total_pay = base_salary
        return total_pay

print("-------------------------")
b1 = abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 = abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native", "Xamarin"])
b2.printInfo()
print("-------------------------")
```

```python
#TASK5


class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, courses, semester):
        total_credits = 0
        total_grade_points = 0
        for course in courses:
            course_code, grade_point = course.split(': ')
            credits = 3
            total_credits += credits
            total_grade_points += float(grade_point) * credits
        gpa = round(total_grade_points / total_credits, 2)
        self.grades[semester] = {tuple(courses): gpa}

    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        for semester, courses in self.grades.items():
            print(f'Courses taken in {semester}:')
            for course, gpa in courses.items():
                print(f'{course[0]}')
                print(f'GPA: {gpa}')

# Write your code here
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0',
'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'],
'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7',
'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
#Task6

class Store:
    def __init__(self, balance):
        self.balance = balance
        self.items = {}
        self.total_items = 0
        print("New branch created!")

    def viewAllItems(self):
        if not self.items:
            print("There are no items in your inventory")
        else:
            print("All Items:", ", ".join(self.items.keys()))

    def viewAllItemDetails(self):
        if not self.items:
            print("There are no items in your inventory")
        else:
            print(self.items)

    def add_item(self, item):
        name, stock, buying_price, selling_price = item
        if name in self.items:
            print(f"{name} already exists in your inventory")
        else:
            self.items[name] = {"stock": stock, "buying_price": buying_price, "selling_price": selling_price}
            self.balance -= stock * buying_price
            self.total_items += stock
            print(f"Item added: {name}")
            print(f"Current Balance: {self.balance}")

    def sell_item(self, name, quantity):
        if name not in self.items:
            print(f"Sorry! {name} is not available in your inventory")
        elif self.items[name]["stock"] < quantity:
            print(f"Sorry! {name} is not available at your desired quantity. Currently we have: {self.items[name]['stock']}")
        else:
            self.items[name]["stock"] -= quantity
            self.balance += quantity * self.items[name]["selling_price"]
            print(f"Current Balance: {self.balance}")
            print(f"{quantity} {name} sold successfully")

    def restock_item(self, name, quantity):
        if name not in self.items:
            print(f"Sorry! {name} is not available in your inventory")
        else:
            self.items[name]["stock"] += quantity
            self.balance -= quantity * self.items[name]["buying_price"]
            print(f"Restocked item: {name}, Current Stock: {self.items[name]['stock']}")
            print(f"Current Balance: {self.balance}")

print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
