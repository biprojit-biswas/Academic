```python
#TASK-01

BMI = 0
def BMI_calculator(height,weight):
  meterHeight = height / 100
  global BMI
  BMI = round(weight / (meterHeight**2),1)
  if BMI < 18.5:
    return "Underweight"
  elif 18.5<= BMI <= 24.9:
    return "Normal"
  elif 25<= BMI <= 30:
    return "Overweight"
  # elif BMI > 30:
  else:
    return "Obese"

weight= int(input("Please, Enter your weight in Kg: "))
height=int(input("Please, Enter your heightin cm: "))
result = BMI_calculator(weight,height)
print("Score is",BMI,"."+"You are",result)
```

```python
#TASK-03

def replace_domain(checked,newDomain,oldDomain = "kaaj.com"):
  for i in range(len(checked)):
    if checked[i] == '@':
      part1 = checked[:i+1]
      domain = checked[i+1:]

  if newDomain == domain:
    print("Unchanged:",checked)
  else:
    print("Changed:",part1+newDomain)

replace_domain("bob@sheba.xyz", "sheba.xyz")
replace_domain("alice@kaaj.com", "sheba.xyz","kaaj.com")
```

```python
#TASK-04

def checkPalindrom(word):
  n=""
  new_word=""

  for i in word:
    if i==" ":
      continue
    else:
      n+=i

  for i in range(-1,-(len(n)+1), -1):

    new_word+=n[i]


  if new_word==n:
    print("Palindrome")

  else:
    print("Not a Palindrome")


word =input('Please enter your word: ')
checkPalindrom(word)
```

```python
#TASK-04

def palindrome(word):
  string = ''
  for i in word:
    if 65<=ord(i)<=90 or 97<=ord(i)<= 122:
      string += i
  forwardString = string[:]
  backwardString = string[-1:-len(string)-1:-1]

  if forwardString == backwardString:
    print("Palindrome")
  else:
    print("Not a palindrome")

palindrome('nurses run')
```
