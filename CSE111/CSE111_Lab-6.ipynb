```python
#TASK1

class Student:
  id = 0
  def __init__(self, name, dept, age, cgpa):
    self.name = name
    self.dept = dept
    self.age = age
    self.cgpa = cgpa
    Student.id += 1

  def showDetails(self):
    print(f"ID: {self.id}")
    print(f"Name: {self.name}")
    print(f"Department: {self.dept}")
    print(f"Age: {self.age}")
    print(f"CGPA: {self.cgpa}")

  @classmethod
  def from_String(cls, details):
    name, dept, age, cgpa = details.split("-")
    return cls(name, dept, int(age), float(cgpa))

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()
print("The difference between an instance variable and a class variable lies in their scope and usage. Instance variables are tied to a specific object instance of a class and are unique to each instance, while class variables are shared among all instances of the class and are defined at the class level. Instance variables are used when the variables are expected to change significantly across instances, whereas class variables are used when the variables are consistent across instances or when variable initialization is required")
print("The difference between an instance method and a class method lies in their behavior and usage. An instance method is a method that is called on an object instance and can access and modify the instance's state through the self parameter. On the other hand, a class method is a method that is called on the class itself and can modify the class state through the cls parameter. Class methods are shared among all instances of the class, while instance methods are specific to each instance")

```

```python
#TASK2


class Passenger:
  count = 0
  base_fare = 450

  def __init__(self, name):
    self.name = name
    self.bag_weight = 0
    self.additional_fare = 0
    Passenger.count +=1

  def set_bag_weight(self, weight):
    self.bag_weight = weight
    if 20 < weight <= 50:
      self.additional_fare = 50
    elif weight > 50:
      self.additional_fare = 100

  def printDetail(self):
    total_fare = self.base_fare + self.additional_fare
    print(f"Name: {self.name}")
    print(f"Bus fare: {total_fare} taka")

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#task4

class NikeBangladesh:
    branches, stock, sold = [], {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}, 0

    def __init__(self, branch):
        self.branch, self.branches_stock, self.branch_sold = branch, {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}, 0
        NikeBangladesh.branches.append(branch)

    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:\nBranches Opened:", cls.branches, "\nCurrently Stocked\n", cls.stock, "\nSold:", cls.sold)

    def details(self):
        print(f"{self.branch} outlet:\nProducts Currently Stocked:\n", self.branches_stock, "\nSold:", self.branch_sold)

    def restockProducts(self, products):
        for p, q in products.items():
            self.branches_stock[p] += q
            NikeBangladesh.stock[p] += q

    def productSold(self, products):
        for p, q in products.items():
            self.branches_stock[p] -= q
            NikeBangladesh.stock[p] -= q
            self.branch_sold += q
            NikeBangladesh.sold += q

print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts({"Air Jordan": 1200, "Cortez": 200, "Zoom Kobe": 200})
chittagong.restockProducts({"Air Jordan": 1000, "Cortez": 250, "Zoom Kobe": 100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan": 760, "Cortez": 90})
chittagong.productSold({"Air Jordan": 520, "Zoom Kobe": 70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
#task6

class SultansDine:
    total_branch = 0
    total_sell = 0
    branch_list = []

    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.total_branch}")
        print(f"Total Sell: {cls.total_sell} Taka")

    def __init__(self, branch):
        self.branch = branch
        self.quantity = 0
        self.sell = 0
        SultansDine.total_branch += 1
        SultansDine.branch_list.append({'branch': self.branch, 'sell': self.sell})

    def sellQuantity(self, quantity):
        self.quantity = quantity
        if self.quantity < 10:
            self.sell = self.quantity * 300
        elif self.quantity < 20:
            self.sell = self.quantity * 350
        else:
            self.sell = self.quantity * 400

        SultansDine.total_sell += self.sell
        for branch in SultansDine.branch_list:
            if branch['branch'] == self.branch:
                branch['sell'] = self.sell

    def branchInformation(self):
        print(f"Branch Name: {self.branch}")
        print(f"Branch Sell: {self.sell} Taka")
        branch_sell_percentage = (self.sell / SultansDine.total_sell) * 100
        print(f"Branch consists of total sell's: {branch_sell_percentage:.2f}%")



SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()


```
