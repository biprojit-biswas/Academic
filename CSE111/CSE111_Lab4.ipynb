```python
class Student:
    def __init__(self, name, cgpa, credits=9, department="CSE"):
        self.name = name
        self.cgpa = cgpa
        self.credits = credits
        self.department = department
        self.scholarship = "No scholarship"

    def checkScholarshipEligibility(self):
        if self.cgpa >= 3.5 and self.credits > 10:
            if self.cgpa >= 3.7:
                self.scholarship = "Merit-based scholarship"
                print(f"{self.name} is eligible for Merit-based scholarship.")
            else:
                self.scholarship = "Need-based scholarship"
                print(f"{self.name} is eligible for Need-based scholarship.")
        else:
            print(f"{self.name} is not eligible for scholarship.")

    def showDetails(self):
        print(f"Name: {self.name} \nDepartment: {self.department} \nCGPA: {self.cgpa} \nNumber of Credits: {self.credits} \nScholarship Status: {self.scholarship}")


print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
class Foodie:
    menu = {
        'Chicken Lollipop': 15,
        'Beef Nugget': 20,
        'Americano': 180,
        'Red Velvet': 150,
        'Prawn Tempura': 80,
        'Saute Veg': 200
    }

    def __init__(self, name):
        self.name = name
        self.cart = []
        self.total_spent = 0

    def show_orders(self):
        if len(self.cart) == 0:
            return f"{self.name} has 0 item(s) in the cart.\nItems: []\nTotal spent: {self.total_spent}."
        else:
            return f"{self.name} has {len(self.cart)} item(s) in the cart.\nItems: {self.cart}\nTotal spent: {self.total_spent}."

    def order(self, *items):
        for item in items:
            food, quantity = item.split('-')
            food = food.strip()
            quantity = int(quantity)

            if food in self.menu:
                price = self.menu[food]
                total_price = price * quantity
                self.cart.append(food)
                self.total_spent += total_price
                print(f"Ordered - {food}, quantity - {quantity}, price (per Unit) - {price}.\nTotal price - {total_price}")

    def pay_tips(self, tips=0):
        if tips > 0:
            self.total_spent += tips
            print(f"Gives {tips}/- tips to the waiter.")
        else:
            print("No tips to the waiter.")

##driver code
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3', 'Beef Nugget-6', 'Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())

f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6', 'Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
class Department:
    def __init__(self, dept='ChE Department', sect=5):
        self.dept = dept
        self.students = []
        self.sections = sect
        self.average = 0
        self.count = 0
        self.total_students_of_dept = 0
        self.average_students_of_dept = 0
        print(f"The {self.dept} has {self.sections} sections.")


    def add_students(self, *nums):
        for student in nums:
            self.students.append(student)
        if len(self.students) == self.sections:
            self.average = round((sum(self.students) / self.sections), 2)
            print(f"The {self.dept} has an average of {self.average} students in each section.")
        else:
            print(f"The {self.dept} doesn't have {len(nums)} sections.")

    def merge_Department(self, *departments):
        merged_departments = []
        total_students_of_dept = 0
        count = 0

        for dept in departments:
            total_students_of_dept += dept.average * dept.sections
            count += dept.sections
            merged_departments.append(dept.dept)

        merged_departments_str = "\n".join(merged_departments)
        print(f"{merged_departments_str} is merged to {self.dept}.")

        self.total_students_of_dept = total_students_of_dept
        self.count = count
        self.average_students_of_dept = round(self.total_students_of_dept / (self.sections + self.count), 2)

        return f"Now the {self.dept} has an average of {self.average_students_of_dept} students in each section."


d1 = Department()
print('1-----------------------------------')
d2 = Department('MME Department')
print('2-----------------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------------')
print(mega.merge_Department(d3))



```

```python
class Department:
    all_departments = []

    def __init__(self, name='ChE Department', sections=5):
        self.name = name
        self.sections = sections
        self.students = [0] * sections
        Department.all_departments.append(self)

    def add_students(self, *num_students):
        if len(num_students) == self.sections:
            self.students = list(num_students)
            average_students = sum(self.students) / self.sections
            print(f'The {self.name} has an average of {average_students:.2f} students in each section.')
        else:
            print(f'The {self.name} doesn\'t have {self.sections} sections.')

    @classmethod
    def merge_Department(cls, *departments):
        mega = Department('Engineering Department', 0)
        total_students = 0
        total_sections = 0

        for department in departments:
            total_students += sum(department.students)
            total_sections += department.sections
            Department.all_departments.remove(department)

        mega.sections = total_sections
        mega.students = [total_students // total_sections] * total_sections

        for department in departments:
            print(f'{department.name} is merged to {mega.name}.')

        mega.add_students(*mega.students)
        return mega.name

# Driver Code
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)

print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Author:
    def __init__(self, name=""):
        self.name = name
        self.books = {}
        self.book_count = 0

    def addBook(self, title, genre):
        if self.name:
            self.book_count += 1
            if genre in self.books:
                self.books[genre].append(title)
            else:
                self.books[genre] = [title]
        else:
            print("A book cannot be added without an author name")

    def setName(self, name):
        self.name = name

    def printDetail(self):
        print("Number of Book(s):", self.book_count)
        print("Author Name:", self.name)
        for genre, titles in self.books.items():
            print(f"{genre}: {', '.join(titles)}")

# Do not change the following lines of code.
a1 = Author()
print("===============================")
a1.addBook("Ice", "Science Fiction")
print("===============================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("===============================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon","Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("===============================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("===============================")
a2.printDetail()
print("===============================")
```
