## **Task 1: Zigzag Walk**

```python
def forward(arr, row, col):
  for i in range(row):
    if arr[i][col] != 0:
      print(arr[i][col], end=" ")
  print()

def backward(arr, row, col):
  for i in range(row-1, -1, -1):
    if arr[i][col] != 0:
      print(arr[i][col], end=" ")
  print()

def walk_zigzag(floor):

  row, col = np.shape(floor)

  result = np.array([[0]*col]*row)
  for i in range(row):
    for j in range(col):
      if (i+j) % 2 == 0:
        result[i][j] = floor[i][j]

  for k in range(col):
    if k % 2 ==0:
      forward(result, row, k)
    else:
      backward(result, row, k)



floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )



print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

## **Task 3: Matrix Manipulation**

```python
def reverse_Matrix(matrix):
  #TO DO
  row,col = matrix.shape

  for i in range(row//2):
    for j in range(col):
      matrix[i] [j] ,matrix[row-1-i][j] = matrix[row-1-i][j] , matrix[i][j]

  for i in range(row):
    for j in range(col//2):
      matrix[i] [j] ,matrix[i][col-1-j] = matrix[i][col-1-j] , matrix[i][j]

  return matrix


matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

## **Task 5: Matrix Compression**

```python
def compress_matrix(mat):
    #TO DO
    row,col= mat.shape

    m=row-2
    n=col-2

    one_D= np.zeros(row, dtype=int)
    comp= np.zeros((m,n), dtype=int)

    x=0
    for i in range(0,row,2):
      for j in range(0,col,2):
        sum = mat[i][j] + mat [i][j+1] + mat[i+1][j]+ mat[i+1][j+1]
        one_D[x]=sum
        x+=1

    p=0
    for k in range(m):
        for l in range(n):
          comp[k][l] = one_D[p]
          p+=1


    return comp


matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

## **Task 6: Game Arena**

```python
def play_game(arena):
  #TO DO
  row,col= arena.shape

  count=0

  for i in range(row):
    for j in range(col):
      if arena[i] [j]!=0 and arena[i][j]% 50== 0:

       if i-1>=0:
        if arena[i-1][j]==2:
          count+=1

        if i +1 <row-1:
          if arena [i+1][j] ==2:
            count+=1

        if j+1 <= col-1:
          if arena[i][j+1] ==2:
            count+=1

        if j-1>=0:
          if arena[i][j-1]==2:
            count+=1

  points= count * 2

  if points < 10:
    print(f"Points Gained: {points}. Your Team is out")
  else:
    print(f"Points Gained: {points}. Your team has survived the game.")



arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
