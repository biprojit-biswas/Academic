```python
class Node:
  def __init__(self, elem, next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.top = None

  def push(self, elem):
    n = Node(elem)
    if self.top == None:
      self.top = n
    else:
      n.next = self.top
      self.top = n

  def pop(self):
    if self.top == None:
      print("Underflow")
    else:
      r = self.top
      self.top = self.top.next
      r.next = None
      return r

  def peek(self):
    if self.top != None:
      print(self.top.elem)
    else:
      print("Underflow")

  def isEmpty(self):
    return self.top == None

  def reverse(self):
    st=Stack()

    while self.peek() != None:
      i=self.pop()
      st.push(i)
    return st


```

```python
import numpy as np
deck_arr = np.array([1, 5, 7, 10, 12, 4, 8, 9, 3, 1, 5, 6])
n=len(deck_arr)

def make_deck(arr):
  deck=Stack()
  for i in range((len(arr)-1),-1,-1):
    deck.push(arr[i])
  return deck



def simulate(deck, p1, p2):
  c=0
  while c<n:
    if c%2==0:
      sum=0
      if deck.peek()!=None:
        x=deck.pop()
        p1.push(x)
        sum+=x
        if sum==21:
          return p1
        elif sum>21:
          return p2

    elif c%2!=0:
      sum=0
      if deck.peek()!=None:
        y=deck.pop()
        p2.push(y)
        sum+=y
        if sum==21:
          return p2
        elif sum>21:
          return p1

    c+=1



def print_pick_order(p):
  m= p.reverse()
  while m.peek() !=None:
    x=m.pop()
    print(x, end=' ')

```

```python
# Driver code


p1 = Stack()
p2 = Stack()

deck = make_deck(deck_arr)

winner = simulate(deck, p1, p2) # should print: "Winner: P1"
print_pick_order(winner) # should print 6 1 9 8 4

deck = make_deck([5, 11, 6, 2, 7, 3, 10, 9])
winner = simulate(deck, p1, p2) # should print: "Winner: P2"
print_pick_order(winner) # should print 11 2 3

deck = make_deck([5, 11, 6, 2, 7, 3, 3, 9])
winner = simulate(deck, p1, p2) # should print: "Winner: P1"
print_pick_order(winner) # should print 5 6 7 3
```
