```python
#Reverse an array

def reverseArray(a):
  b=[0]*len(a)
  i=0
  j= len(a)-1

  while j >=0:
    b[i]=a[j]
    j-=1
    i+=1
  return b


a=[10,20,30,40,50]

a= reverseArray(a)
print(a)
```

```python
#Reverse array

import numpy as np

def reverseArray(a):

  b= np.zeros(len(a))
  i=0
  j= len(a)-1

  while j >=0:
    b[i]=a[j]
    j-=1
    i+=1
  return b

a=[10,20,30,40,50]

a=reverseArray(a)

print(a)
```

```python
#in place

#Reverse an array

def reverseArray(a):

  i=0
  j= len(a)-1

  while j >= (len(a)//2):

    ex=a[i]
    a[i]=a[j]
    a[j] = ex
    j-=1
    i+=1
  return a


a=[10,20,30,40,50,60]

a= reverseArray(a)

print(a)
```

```python
#right shifting

a=[10,20,30,40,50]

i=len(a)-1

while i >=1:
  #t= a[i]
  a[i]= a[i-1]

  #a[0]=0
  i-=1
a[0]=0

print(a)
```

```python
#right shifting for loop

a=[10,20,30,40,50]

for i in range(len(a)-1,0,-1):
  a[i]= a[i-1]

a[0]=0

print(a)
```

```python
#right rotate


a=[10,20,30,40,50]

i=len(a)-1
temp= a[i]

while i >=1:

  a[i]= a[i-1]
  i-=1

a[0]=temp

print(a)
```

```python
#right rotate for loop


a=[10,20,30,40,50]

temp= a[len(a)-1]

for i in range(len(a)-1,0,-1):
  a[i]= a[i-1]

a[0]=temp

print(a)
```
