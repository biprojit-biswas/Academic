```python
#
a=[10,20,30,40,50]

def rightShift(a):

  i= len(a)-1
  while i>=1:
    a[i] = a[i-1]
    i-=1

  a[0]=0
  return a

a=rightShift(a)
print(a)
```

```python
#
a=[10,20,30,40,50]

def rightRotate(a):

  t= a[len(a)-1]

  i= len(a)-1
  while i>=1:
    a[i] = a[i-1]
    i-=1

  a[0]= t
  return a

a=rightRotate(a)
print(a)
```
