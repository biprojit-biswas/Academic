```python
# Question 1

arr = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

def calculateMean(arr):
    total = 0
    size = 0
    for i in range(len(arr)):
        if arr[i] != None:
            total += arr[i]
            size += 1

    return total / size

print(calculateMean(arr))
```

```python
# Question 2

import math

def StandardDeviation(arr):
    mean = calculateMean(arr)

    total = 0
    for i in range(len(arr)):
        total += (arr[i] - mean)**2

    std = math.sqrt((total/(len(arr)-1)))
    return std

arr = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

a=StandardDeviation(arr)
print("% .2f" %a)
```

```python
# Question 3

def leftShift(arr, n):
    for j in range(n, len(arr)-1):
        arr[j] = arr[j+1]
    arr[len(arr)-1] = 0
    return arr


def newArray(arr):

    mean = calculateMean(arr)
    std = StandardDeviation(arr)

    validValue = 0
    for i in range(len(arr)):
        if arr[i] > (mean + std * 1.5) or arr[i] < (mean - std * 1.5):
            validValue += 1

    new_arr = np.zeros(validValue, dtype=int)
    idx = 0
    for i in range(len(arr)):
        if arr[i] > (mean + std * 1.5) or arr[i] < (mean - std * 1.5):
            new_arr[idx] = arr[i]
            idx += 1

    return new_arr

arr = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

print(newArray(arr))
```

```python
#Test 01: Merge Lineup:


def mergeLineup(pokemon_1, pokemon_2):
  result = [None]*len(pokemon_1) #this array stores the resulting array
  a = 0
  b = len(pokemon_2)-1
  while a<len(pokemon_1):
    if pokemon_1[a] !=None and  pokemon_2[b] != None:
        result[a] = pokemon_1[a] + pokemon_2[b]
    elif pokemon_2[b] ==None :
      result[a] = pokemon_1[a]
    elif pokemon_1[a] ==None :
      result[a] = pokemon_2[b]

    a += 1
    b -= 1

  return np.array(result)

print("///  Test 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'{returned_value}') # This should print [12, 3, 28, -8, 5]


pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'{returned_value}') # This should print [4,17,6,27,2]
```

```python
#TASK-02 Discard Cards:

def discardCards(cards, t):
  flag = True
  i = 0

  while i < len(cards) :
    if cards[i] == t :
      if flag:
        for j in range(i, len(cards)-1, 1) :
          cards[j] = cards[j+1]
        i -= 1
      flag = not flag
    i +=  1

  return cards

cards=[1,3,7,2,5,2,2,2,0]
#Function Call:
print(discardCards(cards, t = 2))
```

```python
#TASK-03 DUBER Fare Splitting:

def findGroups(money, fare):

  c = 1
  for i in range(len(money)) :
    sum = 0
    if money[i] == fare :
      print("Group", c, ": ", money[i])
      money[i] = 0
      continue

    for j in range(i+1, len(money)) :
      if money[i] + money[j] == fare :
        print("Group",c,": ", money[i], money[j])
        money[i], money[j] = 0, 0
        c += 1

  for k in range(len(money)) :
    if money[k] != 0 :
      print("Ungrouped :", end=" ")
      for k in range(len(money)) :
        if money[k] != 0 :
          print(money[k], end=" ")
      break

findGroups( [60, 150, 60, 30, 120, 30], 180 )
```

```python
# Task 04: Get Those Hobbies

def Array_count(arr) :
  c = 0
  for i in range(len(arr)) :
    if arr[i] != 0 :
      c += 1
  return c

def analyzeHobbies(* participants): #(* arguments) is used for variable number of parameters
  count = 0
  for i in participants :
    count += len(i)

  box = np.zeros(count, dtype = object)
  c = 0
  for arr in participants :
    for i in range(len(arr)) :
      box[c] = arr[i]
      c += 1

  new_box = np.zeros(count, dtype = object)
  temp = np.zeros(count, dtype = int)

  p = 0
  for i in range(len(box)) :
    temp[i] = 1
    if box[i] == 0 :
      continue
    for j in range(i+1, len(box)) :
      if box[i] == box[j] :
        temp[i] += 1
        box[j] = 0

    new_box[p] = box[i]
    p += 1

  size = Array_count(new_box)
  temp_arr = np.zeros(size, dtype = object)
  for i in range(len(new_box)) :
    if new_box[i] != 0 :
      temp_arr[i] = new_box[i]

  print("Unique Activities in the Town : ")
  print(temp_arr)
  for i in range(len(temp_arr)) :
    print(f"{temp[i]} participant(s) like(s) {temp_arr[i]}.")


print("///  Task 04: Get Those Hobbies  ///")
participant_1 = np.array( ["Hiking", "Reading", "Photography", "Cooking"])
participant_2 = np.array( ["Reading", "Hiking", "Painting"])
participant_3 = np.array( ["Hiking", "Cooking", "Photography"])
print(f'Task 4:')
analyzeHobbies(participant_1, participant_2, participant_3) #This should print

#Unique Activities in the Town:
#['Photography', 'Painting', 'Cooking', 'Reading', 'Hiking']

#Statistics:
#2 participant(s) like(s) Photography.
#1 participant(s) like(s) Painting.
#2 participant(s) like(s) Cooking.
#2 participant(s) like(s) Reading.
#3 participant(s) like(s) Hiking.



participant_1 = np.array( ["Gardening", "Traveling"])
participant_2 = np.array( ["Singing", "Gardening", "Painting"])
print(f'Task 4:')
analyzeHobbies(participant_1, participant_2) #This should print

#Unique Activities in the Town:
#[Gardening, Traveling, Singing, Painting]

#Statistics:
#2 participant(s) like(s) Gardening.
#1 participant(s) like(s) Traveling.
#1 participant(s) like(s) Singing.
#1 participant(s) like(s) Painting.
```

```python
import numpy as np

def generate_look_and_say_sequence(arr):
    result = [0]*100


    count = 1
    current_digit = arr[0]

    for i in range(1, len(arr)):
        if arr[i] == current_digit:

            count += 1
        else:

            result.append(count)
            result.append(current_digit)

            # Reset count and update current digit
            count = 1
            current_digit = arr[i]


    result.append(count)
    result.append(current_digit)


    result_array = np.array(result)

    return result_array

# Example usage:
input_array = np.array([1,3,1,1,2,2,2,2,1])
next_sequence = generate_look_and_say_sequence(input_array)
print(next_sequence)
```

```python
import numpy as np

def look_and_say_sequence(arr):
    result = []

    i = 0
    while i < len(arr):
        count = 1
        while i + 1 < len(arr) and arr[i] == arr[i + 1]:
            i += 1
            count += 1
        result.append(count)
        result.append(arr[i])
        i += 1

    return np.array(result[:100])

# Example usage:
input_array = np.array([1, 1, 2, 1])
next_sequence = look_and_say_sequence(input_array)

print("Input Array:", input_array)
print("Next Sequence:", next_sequence)
```
