#Zigzag Walk

```python
def forward(arr, row, col):
  for i in range(row):
    if arr[i][col] != 0:
      print(arr[i][col], end=" ")
  print()

def backward(arr, row, col):
  for i in range(row-1, -1, -1):
    if arr[i][col] != 0:
      print(arr[i][col], end=" ")
  print()

def walk_zigzag(floor):

  row, col = np.shape(floor)

  result = np.array([[0]*col]*row)
  for i in range(row):
    for j in range(col):
      if (i+j) % 2 == 0:
        result[i][j] = floor[i][j]

  for k in range(col):
    if k % 2 ==0:
      forward(result, row, k)
    else:
      backward(result, row, k)



floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#ATM's Triangle

```python
def sumofrow(arr, row, col):
    store = []*col
    total = arr[row][0]
    store[0] = total
    l = 1
    for k in range(1, col):
        if arr[row][k] != 0 and arr[row][k+1] != 0:
            total += arr[row][k]
            store[l] = total
            l += 1
    return store
def generate_pascal_triangle(n):
    pascal_triangle = np.zeros((n, n), dtype=int)

    for i in range(n+1):
        for j in range(i+1):
            if j == i :
                pascal_triangle[i-1][j-1] = i
                pascal_triangle[i-1][0] = i

    sum = 0
    j = 1
    for i in range(2, n) :
        sum= pascal_triangle[i-1][0]
        while i > j:
            sum += pascal_triangle[i-1][j]

            pascal_triangle[i][j] = sum
            j+= 1
        j = 1

    return pascal_triangle

def print_pascal_triangle(pascal_triangle):

    row, col = np.shape(pascal_triangle)
    for i in range(row):
        for j in range(col):
            if pascal_triangle[i][j] != 0:
                print(pascal_triangle[i][j], end = " ")
        print()


n = 5
pascal_triangle = generate_pascal_triangle(n)
print_pascal_triangle(pascal_triangle)
```
