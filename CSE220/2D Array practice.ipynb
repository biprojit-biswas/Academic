```python
def calculate_points(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    points = 0

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] %50== 0:  # Only consider positions with values > 0
                # Check neighboring cells
                # Up
                if i > 0 and matrix[i - 1][j] == 2:
                    points += 1
                # Down
                if i < rows - 1 and matrix[i + 1][j] == 2:
                    points += 1
                # Left
                if j > 0 and matrix[i][j - 1] == 2:
                    points += 1
                # Right
                if j < cols - 1 and matrix[i][j + 1] == 2:
                    points += 1

    return points

def game_result(matrix):
    points = calculate_points(matrix)
    if points >= 10:
        return f"Points Gained: {points}. Your team has survived the game."
    else:
        return f"Points Gained: {points}. Your team is out."

# Sample Input 1
matrix1 = [
    [0, 2, 2, 0],
    [50, 1, 2, 0],
    [2, 2, 2, 0],
    [1, 100, 2, 0]
]

# Sample Input 2
matrix2 = [
    [0, 2, 2, 0, 2],
    [1, 50, 2, 1, 100],
    [2, 2, 2, 0, 2],
    [0, 200, 2, 0, 0]
]

print(game_result(matrix1))
print(game_result(matrix2))
```

```python

import numpy as np

def compress_matrix(mat):
    #TO DO
    row,col= mat.shape

    m=row-2
    n=col-2

    one_D= np.zeros(row, dtype=int)
    comp= np.zeros((m,n), dtype=int)

    x=0
    for i in range(0,row,2):
      for j in range(0,col,2):
        sum = mat[i][j] + mat [i][j+1] + mat[i+1][j]+ mat[i+1][j+1]
        one_D[x]=sum
        x+=1

    p=0
    for k in range(m):
        for l in range(n):
          comp[k][l] = one_D[p]
          p+=1


    return comp





matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print (matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print (returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```
