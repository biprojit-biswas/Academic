```python
from numpy.core.multiarray import dtype
#Evaluation task(CSE220, Section:28)

#import numpy as np

def calculate_statistics(id):
    #result=[0]*3
    result=np.zeros((3),dtype=int)

    sum=0
    max_value=id[0]
    min_value=id[0]

    for i in range(len(id)-1,-1,-1):
      sum=id[i]+sum

      if id[i]>max_value:
        max_value=id[i]

      if id[i]<min_value:
        min_value=id[i]

    #for i in range(len(id)-1,-1,-1):
      #if id[i]>max_value:
        #max_value=id[i]

    #for i in range(len(id)-1,-1,-1):
      #if id[i]<min_value:
        #min_value=id[i]

    #b[0]=sum
    #b[1]=max_value
    #b[2]=min_value

    for i in range(len(result)):
      if i==0:
        result[i]=sum
      if i==1:
        result[i]=max_value
      if i==2:
        result[i]=min_value
    return result


# Example 1:
number_list_1 = [1,8,1,0,3,4,5,2]
result_1 = calculate_statistics(number_list_1)
print("Example 1 Result:", result_1)  # This should print [24,8,0]

# Example 2:
number_list_2 = [1,9,3,0,3,2,0,3]
result_1 = calculate_statistics(number_list_2)
print("Example 1 Result:", result_1)  # This should print [21,9,0]
```
