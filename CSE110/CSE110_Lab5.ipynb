```python
#5.3

def recursive_summation(start, end):
    # Base case: if start is greater than end, return 0
    if start > end:
        return 0
    # Recursive case: add the current number (start) to the sum of the rest
    else:
        return start + recursive_summation(start + 1, end)

# Example usage:
n = 4
result = recursive_summation(1, n)
print(f"The sum of numbers from 1 to {n} is: {result}")
```

```python
#1.1

def is_prime(n):

    s = 0
    for i in range(1,n+1):

        if n % i == 0:
            s= s+i

    if s % n == 1:
        return True
    else:
        return False


prime_check = is_prime(17)
print(prime_check)
```

```python
#1.2

def is_perfect(number):
    if number <= 0:
        return False

    divisors_sum = 0
    for i in range(1, number):
        if number % i == 0:
            divisors_sum += i

    if  divisors_sum == number:
        return True
    else:
        return False

# Example usage:

n= int(input("Enter your number: "))

result = is_perfect(n)
print(result)
```

```python
#5.3

def special_sum(n):

    sum=0

    for i in range(1,n+1):

        if is_prime(i) == True:
            sum+=i

        elif is_perfect(i) == True:
            sum+=i

    return sum


n =int(input("Enter your number: "))
result = special_sum(n)

print(result)
```

```python
#4.1

def reverse_digits(n):
    # Base case: if n is a single-digit number, print it
    if n < 10:
        print(n)
    else:
        # Print the last digit of n
        print(n % 10)
        # Recursive call with the remaining digits (excluding the last one)
        reverse_digits(n // 10)

# Example usage:
number = 12345
print("Original number:", number)
print("Digits in reverse order:")
reverse_digits(number)
```

```python
#4.2

def power(a, n):
    # Base case: if n is 0, return 1 (anything raised to the power of 0 is 1)
    if n == 0:
        return 1
    # Recursive case: calculate a * a^(n-1)
    else:
        return a * power(a, n - 1)

# Example usage:
base = 3
exponent = 5
result = power(base, exponent)
print(result)
print(f"{base} raised to the power of {exponent} is: {result}")
```

```python
#5.1

def sequence_iterative(N):
    result = 0
    sign = 1  # To alternate between addition and subtraction

    for i in range(1, N + 1):
        result += sign * i
        sign *= -1  # Toggle the sign for the next iteration

    return result

# Example usage:
n = 10
result = sequence_iterative(n)
print(f"The value of y for n={n} is: {result}")
```

```python
#5.2

def sequence_recursive(N):
    # Base case: if N is 1, return 1
    if N == 1:
        return 1

    # Recursive case: calculate the value based on the pattern
    # If N is even, subtract N; if N is odd, add N
    if N % 2 == 0:
        return sequence_recursive(N - 1) - N
    else:
        return sequence_recursive(N - 1) + N

# Example usage
N = 5
result = sequence_recursive(N)
print(f"The value of y for N={N} is: {result}")
```
