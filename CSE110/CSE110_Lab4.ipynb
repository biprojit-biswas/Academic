```python
#Class Evaluation 2

s= input("Enter:")

v="AEIOUaeiou"

vowel_count=0
cons_count=0

for i in s:
  if i != "-" and i != '!':
    if i in v:
      vowel_count += 1
    else:
      cons_count += 1


if vowel_count % 3==0 and cons_count ==0:

  print("Blimey! No Plunder!!")

elif vowel_count ==0 and cons_count % 5 !=0:

  print("Blimey! No Plunder!!")

elif vowel_count % 3==0 and cons_count % 5 ==0:

  print("Aaarr! Me Plunder!!")

else:

  print("Blimey! No Plunder!!")
```

```python
#Task4
text = input()
new_text = ""
count = 0

for i in range(len(text)):
    if not "A"<=text[i]<="Z" and not "a"<=text[i]<="z":
        new_text = new_text + text[i]
    elif count%2==0:
        if "A"<=text[i]<="Z":
            new_text = new_text + text[i]
        else:
            new_text = new_text + chr(ord(text[i])-32)
        count += 1
    elif count%2!=0:
        if "a"<=text[i]<="z":
            new_text = new_text + text[i]
        else:
            new_text = new_text + chr(ord(text[i])+32)
        count += 1

print(new_text)
```
