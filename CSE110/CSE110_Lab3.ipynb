```python
#3

n=int(input("Enter: "))

divisors=""

i=1
while i <= n:
  if n % i ==0:
    divisors+=str(i)
  i+=1

sum=0

if len(divisors) <= 3:
  print(f"{n} is a Prime Number")

elif len(divisors) > 2:
  for j in divisors:
    sum=sum+ int(j)

  if (sum-n)== n:
    print(f"{n} is a Perfect Number")

  else:
    print(f"{n} is not a prime number or perfect number")
```
